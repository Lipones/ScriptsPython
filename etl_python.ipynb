```python
import pandas as pd
```

```python
# Gerar dados aleatórios para ETL
import csv
import random
from datetime import datetime, timedelta

# Número de linhas de dados a serem geradas
num_linhas = 100

# Abre um arquivo CSV para escrita
with open('info_pedidos.csv', mode='w', newline='') as arquivo_csv:
    campo_nomes = ['id_cliente', 'data_compra', 'id_produto', 'quantidade', 'preco']
    escritor_csv = csv.DictWriter(arquivo_csv, fieldnames=campo_nomes)

    # Escreve o cabeçalho no arquivo CSV
    escritor_csv.writeheader()

    # Gera dados aleatórios e escreve no arquivo CSV
    for _ in range(num_linhas):
        id_cliente = random.randint(1, 1000)
        data_compra = (datetime.now() - timedelta(days=random.randint(1, 365))).strftime('%Y-%m-%d')
        id_produto = random.randint(1, 50)
        quantidade = random.randint(1, 10)
        preco = round(random.uniform(10.0, 100.0), 2)

        # Escreve a linha de dados no arquivo CSV
        escritor_csv.writerow({
            'id_cliente': id_cliente,
            'data_compra': data_compra,
            'id_produto': id_produto,
            'quantidade': quantidade,
            'preco': preco
        })

print("Dados aleatórios gerados e salvos em info_pedidos.csv")
```

```python
# Nome do arquivo de entrada e de saída
arquivo_entrada = 'info_pedidos.csv'
arquivo_saida = 'info_pedidos_alter.csv'
```

```python
# Lê o arquivo CSV em um DataFrame
df = pd.read_csv(arquivo_entrada, delimiter=',')
df.head()
```

```python
# Calcula o total vendido (quantidade * preco) e adiciona uma nova coluna
def calcula_total_vendido(df):
  df['total_vendido'] = df['quantidade'] * df['preco']
  return df.head()
```

```python
calcula_total_vendido(df)
df.head()
```

```python
def salva_novo_arquivo(dataframe, nome_arquivo_saida, delimitador = ',', index = False):
  """
  Salva um DataFrame em um arquivo CSV.

  Parâmetros:
  - dataframe: O DataFrame a ser salvo.
  - nome_arquivo_saida: O nome do arquivo de saída.
  - delimitador: O caractere de delimitação a ser usado no arquivo CSV (por padrão, ';').
  - index: Se True, inclui o índice no arquivo CSV (por padrão, False).
  """
  dataframe.to_csv(nome_arquivo_saida, sep = delimitador, index = index)
  print(f"Arquivo '{nome_arquivo_saida}' foi salvo com sucesso")
```

```python
salva_novo_arquivo(df, arquivo_saida)
```
